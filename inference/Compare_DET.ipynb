```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
NC_1 = pd.read_csv('test_metrics_nc_clean_res2net_trial1_non_avg.csv')
NC_2 = pd.read_csv('test_metrics_nc_clean_res2net_trial2_non_avg.csv')
NC_6 = pd.read_csv('test_metrics_nc_clean_res2net_small_trial3_non_avg.csv')
NC_7 = pd.read_csv('test_metrics_nc_clean_res2net_small_trial6_non_avg.csv')
NC_8 = pd.read_csv('test_metrics_nc_clean_res2net_small_trial7_non_avg.csv')
# Assuming CSV files have columns named 'FPR' for False Positive Rate and 'FNR' for False Negative Rate
fpr1, fnr1 = NC_1['FAR'], NC_1['FRR']
fpr2, fnr2 = NC_2['FAR'], NC_2['FRR']
fpr6, fnr6 = NC_6['FAR'], NC_6['FRR']
fpr7, fnr7 = NC_7['FAR'], NC_7['FRR']
fpr8, fnr8 = NC_8['FAR'], NC_8['FRR']

# Plot DET curves
plt.figure(figsize=(10, 8))
plt.plot(fpr1, fnr1, label='Non_chunked_256_mel_bins', marker='o')
plt.plot(fpr2, fnr2, label='Non_chunked_128_mel_bins', marker='s')
plt.plot(fpr6, fnr6, label='Non_chunked_128_mel_bins_small', marker='.')
plt.plot(fpr7, fnr7, label='Non_chunked_128_mel_bins_small_2', marker='.')
plt.plot(fpr8, fnr8, label='Non_chunked_128_mel_bins_new', marker='.')
plt.xlabel('False Positive Rate (FAR)')
plt.ylabel('False Negative Rate (FNR)')
plt.title('DET Curve Comparison')
plt.legend()
plt.xlim(0, 0.2)
plt.ylim(0, 1)


plt.grid(True)
plt.tight_layout()
plt.savefig('DET_Curve_test_clean_Comparison_mel_bins_non_chunked_small_vs_big.png', dpi=300)
# Show the plot
plt.show()
```
