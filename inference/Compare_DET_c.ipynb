```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV files
NC_1 = pd.read_csv('test_metrics_c_res2net_clean_torch.csv')
NC_2 = pd.read_csv('test_metrics_c_res2net_clean.csv')
# Assuming CSV files have columns named 'FPR' for False Positive Rate and 'FNR' for False Negative Rate
fpr1, fnr1 = NC_1['FAR'], NC_1['FRR']
fpr2, fnr2 = NC_2['FAR'], NC_2['FRR']

# Plot DET curves
plt.figure(figsize=(10, 8))
plt.plot(fpr1, fnr1, label='chunked_torchaudio', marker='o')
plt.plot(fpr2, fnr2, label='chunked_librosa', marker='s')
plt.xlabel('False Positive Rate (FAR)')
plt.ylabel('False Negative Rate (FNR)')
plt.title('DET Curve Comparison')
plt.legend()
plt.xlim(0, 0.2)
plt.ylim(0, 1)


plt.grid(True)
plt.tight_layout()
#plt.savefig('DET_Curve_test_clean_Comparison_mel_bins__chunked_small_vs_big.png', dpi=300)
# Show the plot
plt.show()
```
